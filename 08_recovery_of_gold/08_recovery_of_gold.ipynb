# Описание данных

**Технологический процесс**
* Rougher feed — исходное сырье
* Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
* Xanthate **— ксантогенат (промотер, или активатор флотации);
* Sulphate — сульфат (на данном производстве сульфид натрия);
* Depressant — депрессант (силикат натрия).
* Rougher process (англ. «грубый процесс») — флотация
* Rougher tails — отвальные хвосты
* Float banks — флотационная установка
* Cleaner process — очистка
* Rougher Au — черновой концентрат золота
* Final Au — финальный концентрат золота

**Параметры этапов**
* air amount — объём воздуха
* fluid levels — уровень жидкости
* feed size — размер гранул сырья
* feed rate — скорость подачи

**Наименование признаков**

Наименование признаков должно быть такое:

`[этап].[тип_параметра].[название_параметра]`

Пример: `rougher.input.feed_ag`

Возможные значения для блока `[этап]`:
* rougher — флотация
* primary_cleaner — первичная очистка
* secondary_cleaner — вторичная очистка
* final — финальные характеристики

Возможные значения для блока `[тип_параметра]`:
* input — параметры сырья
* output — параметры продукта
* state — параметры, характеризующие текущее состояние этапа
* calculation — расчётные характеристики

### Эффективность обогащения

Проверим, что эффективность обогащения рассчитана правильно. Вычислим её на обучающей выборке для признака **rougher.output.recovery**.

```python
C = train['rougher.output.concentrate_au']  # доля золота в концентрате после флотации/очистки
F = train['rougher.input.feed_au']          # доля золота в сырье/концентрате до флотации/очистки
T = train['rougher.output.tail_au']         # доля золота в отвальных хвостах после флотации/очистки
```

Эффективность обогащения рассчитывается по формуле: `recovery = 100 * C * (F - T) / (F * (C - T))`

```python
calc_recovery_au = 100 * C * (F - T) / (F * (C - T))
```

Найдите MAE между расчётами и значением признака.

```python
print(mean_absolute_error(train['rougher.output.recovery'], calc_recovery_au))
```

Как видно значение МАЕ очень мало, значит расчеты **rougher.output.recovery** верны.

## Модель

### sMAPE.

Напишим функцию для вычисления итоговой sMAPE.

```python
def smape(target, predict):
    return  100 / len(target) * sum(abs(target - predict) / ((abs(target) + abs(predict)) / 2))
```

```python
def smape_total(rougher, final):
    return 0.25*rougher + 0.75*final
```

<font color='blue'><b>Комментарий ревьюера: </b></font> ✔️\
<font color='green'> `smape` и `smape_total` расчитаны верно.</font>

### Обучение модели

Заполним целевые признаки для тестовой выборки.

```python
test['final.output.recovery'] = full['final.output.recovery']
test['rougher.output.recovery'] = full['rougher.output.recovery']
```

#### Разобьем признаки для обучения

```python
columns_rouger, columns_final = [], []

for row in list(train.columns):
    if 'rougher' in row:
        columns_rouger.append(row)
    else:
        columns_final.append(row)
```

```python
features_train_rougher = train[columns_rouger].drop(columns='rougher.output.recovery')
features_train_final = train[columns_final].drop(columns='final.output.recovery')

tagert_train_rougher = train['rougher.output.recovery']
tagert_train_final = train['final.output.recovery']

features_test_rougher = test[columns_rouger].drop(columns='rougher.output.recovery')
features_test_final = test[columns_final].drop(columns='final.output.recovery')

tagert_test_rougher = test['rougher.output.recovery']
tagert_test_final = test['final.output.recovery']
```

Посчитаем эффективность обогащения чернового и финального концентрата

#### Линейная регрессия

```python
model = LinearRegression() 

smape_linear_rougher = cross_val_score(model, features_train_rougher, tagert_train_rougher, scoring = make_scorer(smape), cv = 5).mean()
smape_linear_final = cross_val_score(model, features_train_final, tagert_train_final, scoring = make_scorer(smape), cv = 5).mean()
print('Эффективность обогащения чернового концентрата:', smape_linear_rougher)
print('Эффективность обогащения финального концентрата:', smape_linear_final)
```

Рассчитаем итоговую метрику для линейной регрессии

```python
smape_total_linear = smape_total(smape_linear_rougher, smape_linear_final)
print(smape_total_linear)
```

#### Дерево решений

```python
best_model_decision_rougher = None
smape_decision_rougher = 10000
for depth in range(1, 6):
    model = DecisionTreeRegressor(random_state=12345, max_depth=depth) 
    result = cross_val_score(model, features_train_rougher, tagert_train_rougher, scoring = make_scorer(smape), cv = 5).mean()
    if result < smape_decision_rougher:
        best_model_decision_rougher = model
        smape_decision_rougher = result
print('Лучшая модель эффективности обогащения чернового концентрата:', best_model_decision_rougher)
print('Эффективность обогащения чернового концентрата:', smape_decision_rougher)
```

```python
best_model_decision_final = None
smape_decision_final = 1000
for depth in range(1, 6):
    model = DecisionTreeRegressor(random_state=12345, max_depth=depth) 
    result = cross_val_score(model, features_train_final, tagert_train_final, scoring = make_scorer(smape), cv = 5).mean()
    if result < smape_decision_final:
        best_model_decision_final = model
        smape_decision_final = result
print('Лучшая модель эффективности обогащения финального концентрата:', best_model_decision_final)
print('Эффективность обогащения финального концентрата:', smape_decision_final)
```

Рассчитаем итоговую метрику для дерева решений
